# 1. Librerias

```python
import pandas as pd
```

# 2. Importamos CSV

```python
ruta_archivo_csv = '../01_data/shows_atrapalo.csv'

df_shows_bruto = pd.read_csv(ruta_archivo_csv)

df_shows_bruto.head(20)
```

# 3. Divido CSV de los shows

**1 - TABLA ARTISTAS:** Hay shows con más de un artista. Anteriormente hicimos un ‘explode()’ para separarlos en diferentes filas, pero cómo puede generar confusión creamos una tabla externa donde recoja los diferentes artistas por show.

**2 - CUENTAS LOCALES:** Cuando tu buscas monólogos en Madrid en Atrápalo hay diferentes tipos de cuentas y shows. Cada entrada puede ser la cuenta del local/teatro, donde te aparecen todos los shows disponibles en el local en las próximas fechas, o cuentas de un show en concreto.

Para que no haya confusión a la hora de interpretar y analizar los datos vamos a separar las cuentas de local en un data frame y CSV diferente.

## 3.1 TABLA ARTISTAS

Creó un nuevo df y su correspondiente CSV únicamente con los artistas por show.

```python
df_artistas = df_shows_bruto[['nombre_show', 'artista']].copy()
df_artistas
```

```python
# Especifico la ruta y el nombre del archivo CSV
ruta_archivo_csv = '../01_data/artistas_shows.csv'

# Guardo el DataFrame en un archivo CSV
df_artistas.to_csv(ruta_archivo_csv, index=False)
```

Ahora quito los artistas de la tabla y elimino los duplicados.

```python
df_shows_bruto.drop(axis = 1, columns= 'artista', inplace = True)
df_shows_bruto
```

```python
# Eliminamos filas duplicadas, ya que había una fila por artistas y hay más de un show con más de un artista
df_shows_bruto.drop_duplicates(inplace = True)
df_shows_bruto
```

## 3.2 CUENTAS LOCALES

Vamos a crear dos data frames diferentes y ponerlos en dos CSVs diferentes, uno de cuentas locales y otro de los shows:

```python
df_cuentas_locales = df_shows_bruto[df_shows_bruto['tipo_show'] == 'cuenta_local'].copy()
df_cuentas_locales
```

```python
# Especifico la ruta y el nombre del archivo CSV
ruta_archivo_csv = '../01_data/cuentas_locales.csv'

# Guardo el DataFrame en un archivo CSV
df_cuentas_locales.to_csv(ruta_archivo_csv, index=False)
```

```python
df_shows = df_shows_bruto[df_shows_bruto['tipo_show'] != 'cuenta_local'].copy()
df_shows
```

```python
# Especifico la ruta y el nombre del archivo CSV
ruta_archivo_csv = '../01_data/shows.csv'

# Guardo el DataFrame en un archivo CSV
df_shows.to_csv(ruta_archivo_csv, index=False)
```
