# 1. Librerias

```python
import pandas as pd
```

# 2. Importamos CSV

```python
ruta_archivo_csv = '../01_data/shows.csv'

df_shows = pd.read_csv(ruta_archivo_csv)

df_shows.head(20)
```

# 3. Coordinamos datos

Queremos unir los datos que tenemos de los locales con el de los shows, y para eso necesitamos que en ambos archivos los nombres de los locales se llamen igual.

Al volcar los datos en Tableau hemos visto que hay dos locales que no coinciden:

```python
df_shows['local'] = df_shows['local'].replace({
    'MadBeerHall': 'Mad Beer Hall',
    'Cafetería Valverde': 'Cervecería Valverde Comedy Club'
})
```

# 4. Guardamos

```python
# Especifico la ruta y el nombre del archivo CSV
ruta_archivo_csv = '../01_data/shows.csv'

# Guardo el DataFrame en un archivo CSV
df_shows.to_csv(ruta_archivo_csv, index=False)
```
